```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Distribution 1 parameters
mean1 = 0.0004585081258346813
var1 = 9.822386091710335e-05
std1 = np.sqrt(var1)

# Distribution 2 parameters
mean2 = -0.05
var2 = 0.0006560654396246523
std2 = np.sqrt(var2)

# Create a combined x-axis range that covers both distributions
x_min = min(mean1 - 4*std1, mean2 - 4*std2)
x_max = max(mean1 + 4*std1, mean2 + 4*std2)
x = np.linspace(x_min, x_max, 1000)

# PDF calculations
pdf1 = norm.pdf(x, loc=mean1, scale=std1)
pdf2 = norm.pdf(x, loc=mean2, scale=std2)

# Plotting
plt.plot(x, pdf1, label=f'Dist 1 ($\mu$={mean1:.6f}, $\sigma^2$={var1:.1e})', color='blue')
plt.plot(x, pdf2, label=f'Dist 2 ($\mu$={mean2:.6f}, $\sigma^2$={var2:.1e})', color='orange')

plt.title("Comparison of Two Normal Distributions")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```
